```python
from PIL import Image
import os
```

```python
# assign directory
directory = '../Tiles/HugeTile51/usefulTiles/tif'
 
# iterate over files in
# that directory
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a tif file
    if os.path.isfile(f) and filename.endswith(".tif"):
        im = Image.open(f) # Can be many different formats.
        pix = im.load()
        clrs = im.getcolors()
        if(clrs != None):
            print("BLANK, Deleting...",f)
            os.remove(f)
        # else:
        #      print("La imagen contiene data",f)

print("!Done...")
```
